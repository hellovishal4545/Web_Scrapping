```python
url ='https://en.wikipedia.org/wiki/List_of_largest_companies_in_India'

page =requests.get(url)

soup= BeautifulSoup(page.text,'html')
```

```python
soup
```

```python
table = soup.find_all('table')[1]
```

```python
table
```

```python
soup.find('table',class_='wikitable sortable')
```

```python
world_tit  = table.find_all('th')
```

```python
world_tit
```

```python
world_tab_tit = [title.text.strip() for title in world_tit]
world_tab_tit
```

```python
df = pd.DataFrame(columns =world_tab_tit)

df
```

```python
column_data = table.find_all('tr')
```

```python
column_data
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data=  [data.text for data in row_data]
    print(individual_row_data)
    
    length = len(df)
    df.loc[length] = individual_row_data
```

```python
df.set_index('Rank')
```

```python
df.head(2)
```

```python
df.groupby('Industry')['Revenue growth'].max().sort_values(ascending=True)
```

```python
df.groupby('Name')['Revenue growth'].max().sort_values(ascending=False)
```
